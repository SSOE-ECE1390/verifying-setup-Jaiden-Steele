# Import Required Libaries

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

# Main Code

```python
# Load Image
image = cv2.imread('Data/Apollo_11_Launch.jpg')

# Show image
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Initialize Crop Variables
x, y, _ = image.shape

crop_x_start = int(x * 0.1)
crop_x_end = int(x * 0.9)
crop_y_start = int(y * 0.1)
crop_y_end = int(y * 0.9)

cropped_image = image[crop_y_start:crop_y_end, crop_x_start:crop_x_end]
plt.imshow(cropped_image)
plt.title('Cropped Image')
plt.axis('off')
plt.show()

# Start Brightness Adjustment
matrix = np.ones(image.shape, dtype="uint8") * 50
dimmed_image = cv2.subtract(image, matrix)
plt.imshow(dimmed_image)
plt.title('Dimmed Image')
plt.axis('off')
plt.show()

# Rotate Image 
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated_image)
plt.title('Rotated Image')
plt.axis('off')
plt.show()

# Save Modified Image
cv2.imwrite('Data/Apollo_11_Launch_Cropped.jpg', cropped_image)
cv2.imwrite('Data/Apollo_11_Launch_Dimmed.jpg', dimmed_image)
cv2.imwrite('Data/Apollo_11_Launch_Rotated.jpg', rotated_image)
```
